```python
from google.colab import drive
from googleapiclient.discovery import build
from google.colab import auth
import os
import json

# Authenticate and build Drive API client
auth.authenticate_user()
drive_service = build('drive', 'v3')

drive.mount('/content/drive')

BASE_PATH = "/content/drive/MyDrive/ai_reels_game"

# Load clip_scores.json
with open(os.path.join(BASE_PATH, "clip_scores.json"), "r") as f:
    clip_scores = json.load(f)

# Search for each file on Drive by name
for clip in clip_scores:
    filename = clip["filename"]

    # Search for the file
    query = f"name = '{filename}'"
    results = drive_service.files().list(
        q=query, fields="files(id, name)"
    ).execute()

    files = results.get("files", [])

    if files:
        file_id = files[0]["id"]
        clip["video_url"] = f"https://drive.google.com/uc?export=download&id={file_id}"
        print(f"✓ {filename} → {file_id[:20]}...")
    else:
        print(f"✗ {filename} → NOT FOUND")

# Save updated clip_scores.json
output_path = os.path.join(BASE_PATH, "clip_scores.json")
with open(output_path, "w") as f:
    json.dump(clip_scores, f, indent=2)

print(f"\n✅ Updated clip_scores.json with video URLs!")
```
